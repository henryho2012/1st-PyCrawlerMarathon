```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json

#ä¸‹è¼‰è³‡æ–™
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response)

#ä¸€æ¬¡å›å‚³ç­†æ•¸
print(len(data))

#åŒ…å«çš„æ¬„ä½
dict.keys(data[0])
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print('ã€Œæ¨™é¡Œã€:', d['title'], 'ã€Œè²¼æ–‡æ™‚é–“ã€:', d['createdAt'], 'ã€Œç•™è¨€äººæ•¸ã€:', d['commentCount'], 'ã€ŒæŒ‰è®šäººæ•¸ã€:', d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

CommentCount, LikeCount, TotalCount = 0, 0, 0
for d in data:
    CommentCount = CommentCount + d['commentCount']
    LikeCount = LikeCount + d['likeCount']
    TotalCount = TotalCount + 1

print('ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€:', CommentCount/TotalCount, 'ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€:', LikeCount/TotalCount)
```

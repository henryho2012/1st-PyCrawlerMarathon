### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find('em').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd

newslist = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    newslist.append([d.find('em').text, d.find_all('a')[-1].text])

df = pd.DataFrame(newslist, columns=['類別', '標題'])
grouped = df.groupby('類別')
grouped.groups
```

### ③ 哪一個類別的文章最多

```python
df2 = pd.DataFrame(df.groupby('類別').size(), columns=['文章數量'])
df2['排名'] = df2.rank(method='min', ascending=False)
df2.sort_values(by=['排名'])
```
